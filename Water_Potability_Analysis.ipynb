**Data Visualization**

```python
plt.figure(figsize=(15, 10))
sns.histplot(data.Potability)
plt.title("Distribution of Unsafe and Safe Water")
plt.show()
```

```python
sns.histplot(data['ph'])
plt.title("Factors affecting quality of water: ph ")
plt.show()
```

```python
sns.scatterplot(data['Hardness'])
plt.title("Factors affecting quality of water: ph ")
plt.show()
```

```python
sns.distplot(data['Solids'])
plt.title("Factors affecting quality of water: Solids ")
plt.show()
```

```python
sns.lineplot(data['Chloramines'])
plt.title("Factors affecting quality of water: Chloramines ")
plt.show()
```

```python
sns.kdeplot(data['Sulfate'])
plt.title("Factors affecting quality of water: Sulfate ")
plt.show()
```

```python
sns.histplot(data['Conductivity'])
plt.title("Factors affecting quality of water: Conductivity ")
plt.show()
```

```python
sns.scatterplot(data['Organic_carbon'])
plt.title("Factors affecting quality of water: Organic Carbon ")
plt.show()
```

```python
sns.distplot(data['Trihalomethanes'])
plt.title("Factors affecting quality of water: Trihalomethanes ")
plt.show()
```

```python
sns.kdeplot(data['Turbidity'])
plt.title("Factors affecting quality of water: Turbidity ")
plt.show()
```
